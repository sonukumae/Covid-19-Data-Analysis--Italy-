```python
# Step 1: Import libraries
import pandas as pd
import os
import matplotlib.pyplot as plt

# Step 2: Define dataset folder
extract_path = r"C:\Users\Sonu Kumar\Downloads\covid_dataset"

# Step 3: List all CSV files
csv_files = os.listdir(extract_path)
print("Available CSV files:", csv_files)

# Step 4: Load datasets into a dictionary
dataframes = {}
for file in csv_files:
    if file.endswith('.csv'):
        path = os.path.join(extract_path, file)
        df = pd.read_csv(path)
        # Convert date column to datetime
        df['data'] = pd.to_datetime(df['data'])
        dataframes[file] = df

# Step 5: Quick look at each dataset
for name, df in dataframes.items():
    print(f"\n{name} - shape: {df.shape}")
    display(df.head())

# Step 6: Plot key trends from national_data.csv
national_df = dataframes['national_data.csv']

plt.figure(figsize=(12,6))
plt.plot(national_df['data'], national_df['nuovi_positivi'], marker='o', label='New Positive Cases')
plt.plot(national_df['data'], national_df['deceduti'], marker='x', label='Deaths', color='red')
plt.title('COVID-19 Trends in Italy')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

# Step 7: Optional - describe statistics
print("\nSummary statistics for national data:")
display(national_df.describe())
```
